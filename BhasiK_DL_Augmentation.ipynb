```python
import random
import cv2
import skimage.io as io
import os
from skimage import img_as_ubyte, filters
from scipy import ndarray
import skimage as sk
import numpy as np


# Define the image augmentation functions

#Function to rotate the image anticlockwise at a random angle
def anticlockwise_rotation(image):
    #between 0 and 180 degrees
    angle = random.randint(0, 180)
    return sk.transform.rotate(image, angle)

# Function to rotate the image clockwise at a random angle
def clockwise_rotation(image):
    #between 0 and 180 degrees
    angle = random.randint(0, 180)
    return sk.transform.rotate(image, -angle)

#Function to flip the image horizontally
def h_flip(image):
    return  np.fliplr(image)
#Function to flip the image vertically
def v_flip(image):
    return np.flipud(image)
#Function to apply a warp shift to the image
def warp_shift(image):
    # generates a random magnitude value between 0 and 0.2 to define how much the image will be shifted.
    magnitude = random.uniform(0, 0.2)
    #randomly selects either -1 or 1 to determine the direction of the shift.
    direction = random.choice([-1, 1])
    #creates an affine transform object using the selected direction and magnitude to specify the translation.
    transform = sk.transform.AffineTransform(translation=(direction*magnitude, direction*magnitude))
    #applies the affine transform to the input image and returns the warped image with a wrapping mode, which extends the image edges to fill the empty space created by the transformation.
    return sk.transform.warp(image, transform, mode='wrap')
# Function to add random noise to the image
def add_noise(image):
    noisy_image = sk.util.random_noise(image)
    return np.clip(noisy_image, 0, 1)
#Function to apply gaussian blur to the image

def blur_image(image):
    sigma = random.uniform(0, 5)
    return filters.gaussian(image, sigma=sigma, multichannel=True)
```

```python

# Define the paths to the input and output folders
input_folders = ["./dataset/data/split/train/firearm", "./dataset/data/split/train/knife", "./dataset/data/split/test/firearm","./dataset/data/split/test/knife"]
output_folders = ["./dataset/data/augmented/train/firearm", "./dataset/data/augmented/train/knife", "./dataset/data/augmented/test/firearm", "./dataset/data/augmented/test/knife"]

# Define the augmentations to be used
augmentations = [anticlockwise_rotation, clockwise_rotation, h_flip, v_flip, warp_shift, add_noise, blur_image]

# Loop over all the input folders
for folder_idx, input_folder in enumerate(input_folders):
    # Create the corresponding output folder if it doesn't exist
    output_folder = output_folders[folder_idx]
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Loop over all the files in the input folder
    for file_name in os.listdir(input_folder):
        # Only process JPG files
        if not file_name.endswith(".jpg"):
            continue

        # Load the image
        image = io.imread(os.path.join(input_folder, file_name))

        # Apply random augmentations to the image
        augmented_images = []
        for i in range(4):
            img = image.copy()
            img = random.choice(augmentations)(img)
            augmented_images.append(img)

        # Save the augmented images in the output folder
        for i, augmented_image in enumerate(augmented_images):
            output_file_name = os.path.splitext(file_name)[0] + "_augmented_" + str(i) + ".jpg"
            augmented_image_uint8 = (augmented_image * 255).astype(np.uint8)
            io.imsave(os.path.join(output_folder, output_file_name), augmented_image_uint8)
```
